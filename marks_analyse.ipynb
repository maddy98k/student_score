# EDA- Exploratory data analysis

## Gender Distribution

```python
ax=sns.countplot(x="Gender",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

*From the above charts we analyse that number of female is slightly higher than male*

*from the above chart we analyse that group C have the higher numbers of others groups*

```python
df.columns
```

## Parent Education

```python
parentedu=df.ParentEduc.value_counts().reset_index()
parentedu
```

```python
ax=sns.barplot(x="ParentEduc",y="count",data=parentedu,saturation=100)
for bars in ax.containers:
    ax.bar_label(bars)
sns.set(rc={'figure.figsize':(12,5)})
plt.title("parent from diffrent background degree's")
```

```python
df.columns
```

```python
gb=df.groupby(["ParentEduc"]).agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
gb
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb,annot=True)
plt.title("Relationship between  Parent Education and student's score  ")


plt.show()
```

*from the above charts we analyse that good education of parents have good impact on children*

```python
gb1=df.groupby(['ParentMaritalStatus']).agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
gb1

```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1,annot=True)
plt.title("Relationship between ParentMaritalStatus and student's score")
plt.show()
```

*from the above charts we analyse that 'ParentMaritalStatus' have no/negligible impact on student education*

# Ethnic Group Distribution

```python
ethenic=df.EthnicGroup.value_counts().reset_index()
epg=ethenic["EthnicGroup"]
epc=ethenic["count"]

plt.pie(epc,labels=epg,autopct="%1.2f%%")
plt.title("distribution of ethnic group")
plt.show()
```

*from the above charts we analyse that group C have more percentage than others *
